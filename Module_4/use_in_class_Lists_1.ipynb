# Slicing

```python
letters[1:3]
```

## A slice is a _copy_

```python
myslice = letters[1:3]
```

## There are defaults if you omit slice values

```python
letters[0:3]
```

```python
letters[3:]
```

```python
letters[:3]
```

```python
letters[:]
```

## You can slice with negative indexes, too...

```python
letters[-3:-1]
```

```python
letters[-5:-2]
```

## When slicing, you can supply a _step_ to skip items

```python
letters[0::2]
```

```python
letters[1::2]
```

```python
letters[0:5:3]
```

## The step can be _negative_, which then goes backwards

```python
letters
```

```python
letters[-1:-4:-1]  # third number is the step
```

```python
letters[::-1]
```
