### 15. The formula for the monthly payment on a loan can be written with this equation…
$payment=principal × \frac{mrate 〖1+mrate〗^n}{〖1+mrate〗^n -1}$

… where `principal` is the amount of the loan, `mrate` is the interest rate per month, and `n` is the number of months over which the loan is to be repaid. The code below contains some blanks (........). Fill in the blanks so that `payment` is assigned the correct monthly payment. (Colab will underline the ........ as syntax errors; ignore the underlining.)


````python

principal = 10000
annual_rate = .0299
mrate = annual_rate / 12
payback_years = 5
n = payback_years * 12

exponent_term = (1 + mrate) ** n
numerator = mrate * exponent_term
denominator = exponent_term - 1

payment = principal * (numerator / denominator)
````
