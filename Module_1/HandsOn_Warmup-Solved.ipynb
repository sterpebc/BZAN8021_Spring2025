## Problem 4
In code, we often want to represent a series of values that go together. In Python, this is called a _list_. Here is a simple list containing three strings:
```
some_list = ['abc', 'def', 'ghi']
```
Every element of a list has a position in the list, called its _index_. In the list above, the 'abc' is at index 0, the 'def' is at index 1, and the 'ghi' is at index 2. As you can see, list indexes start at 0, not 1. If you want to obtain the value at a certain index, you use a special square bracket notation. For example, to obtain the value at index 1:
```
some_list[1]
```
Although the _some_list_ list above consists of all strings, the elements of a list do not have to be of the same data type. Python lets you "mix and match."

````python
import time
import random

def recolor(matrix):
  for row in matrix:
    for index in range(0, len(row)):
      r, g, b = random.choice(colors)
      row[index].style.button_color = f'#{r:02x}{g:02x}{b:02x}'

def set_final(matrix):
  c = ['#d1940f', '#7a1009']
  data = [0b0011100001111000, 0b0010010010000100, 0b0010010010000000,
          0b0010100010000000, 0b0011000010000000, 0b0010100010000000,
          0b0010010010000000, 0b0010010010000100, 0b0011100001111000,
          0b0000000000000000, 0b0000000000000000, 0b0000000000000000,
          0b1110000110010001, 0b1001001001010001, 0b1001001001001010,
          0b1110001001000100, 0b1010001001001010, 0b1001001001010001,
          0b1000100110010001, 0b0000000000000000]
  for rowindex in range(0, len(matrix)):
    bp = data[rowindex]
    for i, b in enumerate(f'{bp:016b}'):
      matrix[rowindex][i].style.button_color = c[int(b)]

colors = [(random.randint(0, 255), random.randint(0, 255), random.randint(0, 255)) for i in range(0, B)]
l1 = widgets.Layout(width='20px', height='20px')

out.clear_output()

rows = []
for rownum in range(0, A//5):
  row = []
  for colnum in range(0, C//5):
    button = widgets.Button(layout=l1)
    r, g, b = random.choice(colors)
    button.style.button_color = f'#{r:02x}{g:02x}{b:02x}'
    row.append(button)
  rows.append(row)

grid = []
for button_row in rows:
  hb = widgets.HBox(button_row)
  grid.append(hb)

vb = widgets.VBox(grid)
recolor_count = 10

with out:
  display(vb)
  while recolor_count > 0:
    recolor(rows)
    time.sleep(1)
    recolor_count = recolor_count - 1

  set_final(rows)
````
